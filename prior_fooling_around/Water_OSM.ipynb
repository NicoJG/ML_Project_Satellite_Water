```python
lat,lon = 51.48998869749565, 7.512899304101557
dlat = 0.01
dlon = 0.01
```

```python
bbox_overpass = tuple((lat-dlat, lon-dlon, lat+dlat, lon+dlon))
```

```python
import requests
import json
```

```python
overpass_url = "http://overpass-api.de/api/interpreter"
overpass_query = f"""
/*
This has been generated by the overpass-turbo wizard.
The original search was:
“waterway”
*/
[out:json][timeout:25];
// gather results
(
  relation["type"="waterway"]{bbox_overpass};
  way["natural"="water"]{bbox_overpass};
  relation["natural"="water"]{bbox_overpass};
);
// print results
out body;
>;
out skel qt;
"""
response = requests.get(overpass_url, 
                        params={'data': overpass_query})
```

```python
print(overpass_query)
```

```python
data = response.json()
print(data)
```

```python
nodes = {elem['id']:{'lat':elem['lat'],'lon':elem['lon']} for elem in data['elements'] if elem['type'] == 'node'}
```

```python
print(nodes)
```

```python
import matplotlib.pyplot as plt
```

```python
for elem in data['elements']:
    if elem['type'] == 'way':
        node_ids = elem['nodes']
        lat = [nodes[node_id]['lat'] for node_id in node_ids]
        lon = [nodes[node_id]['lon'] for node_id in node_ids]
        plt.plot(lon,lat)
        
plt.xlim(bbox_overpass[1],bbox_overpass[3])
plt.ylim(bbox_overpass[0],bbox_overpass[2])
plt.show()
```
